1. Obter DataFrame (Tabela) com as empresas e o valor dos 10 maiores ativos totais

image.png

```python
df_ativo_total = bpa_2021[bpa_2021['CD_CONTA'] == '1']
```

```python
df_ativo_total = df_ativo_total.sort_values('VL_CONTA', ascending=False).head(10)
```

```python
df_ativo_total = df_ativo_total.pivot(index='DENOM_CIA', columns='DS_CONTA', values='VL_CONTA')
```

```python
df_ativo_total = df_ativo_total.sort_values('Ativo Total', ascending=False)
```

```python
df_ativo_total.reset_index(inplace=True)
```

```python
df_ativo_total.columns = ['Empresas','Ativos']
df_ativo_total
```
