## Download dos Formulários de Referência (2021)

```python
url_fre = 'http://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/FRE/DADOS/fre_cia_aberta_2021.zip'
pasta_drive = '/content/drive/MyDrive/MiniCurso - Python/Pratica/Aula 02 - Acessando dados da B3'
wget.download(url_fre, out=pasta_drive)
```

```python
ZipFile(pasta_drive+'/fre_cia_aberta_2021.zip').extractall(pasta_drive+'/fre_2021')
```
