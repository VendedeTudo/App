```python
import yfinance as yf
import plotly.graph_objects as go

import pandas as pd
from datetime import datetime

# Definir formatação do float
pd.set_option('float_format','{:,.2f}'.format)

import warnings
warnings.filterwarnings('ignore')
```

```python
df_contacoes = yf.download('VALE3.SA', start='2019-01-01', end='2022-05-05')
```

```python
df_contacoes
```

```python
fig = go.Figure(data=[go.Candlestick(x=df_contacoes.index,
                open=df_contacoes['Open'],
                high=df_contacoes['High'],
                low=df_contacoes['Low'],
                close=df_contacoes['Close'])])
fig.show()
```

```python
fig = go.Figure([go.Bar(x=df_contacoes.index, y=df_contacoes['Volume'])])
fig.add_hline(y=df_contacoes['Volume'].mean())
fig.show()
```
