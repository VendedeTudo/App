```python
salario = int(input('Salário: '))

if salario > 1250:
  aumento = salario + salario * (0.1)
  print(aumento)
else:
  aumento = salario + salario * (0.15)
  print(aumento)
```

```python
contador = 1
final = 5
while contador <= final:
  salario = int(input('Salário: '))

  if salario > 1250:
    aumento = salario + salario * (0.1)
    print(aumento)
  else:
    aumento = salario + salario * (0.15)
    print(aumento)

  contador += 1
```

```python
for i in range(1,4):
  salario = int(input('Salário: '))

  if salario > 1250:
    aumento = salario + salario * (0.1)
    print(aumento)
  else:
    aumento = salario + salario * (0.15)
    print(aumento)
```
