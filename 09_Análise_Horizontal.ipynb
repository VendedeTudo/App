```python
bpa_engie_varios_anos = pd.DataFrame()

for ano in range(2021,2022):
  # Ler informações de arquivo (.csv) específico
  bpa = pd.read_csv(pasta_drive+f'/dfp_cia_aberta_BPA_con_{ano}.csv', sep=';', encoding='ISO-8859-9', decimal=',')

  # Filtrar informações da coluna ORDEM_EXERC == ÚLTIMO
  bpa = bpa[bpa['ORDEM_EXERC'] == 'ÚLTIMO']
  bpa.head()

  # Filtrar empresa por CNPJ
  bpa_engie = bpa[ bpa['CNPJ_CIA'] == '02.474.103/0001-19']

  # Filtrar apenas as colunas que serão utilizadas
  bpa_engie = bpa_engie[['DT_REFER','CD_CONTA','DS_CONTA','VL_CONTA']]

  # Converter valores numéricos
  bpa_engie['VL_CONTA'] = pd.to_numeric(bpa_engie['VL_CONTA'])

  # Converter datas
  bpa_engie['DT_REFER'] = pd.to_datetime(bpa_engie['DT_REFER'])

  # Criar coluna ANO
  bpa_engie['ANO'] = bpa_engie['DT_REFER'].dt.year

  # Reduzir as contas
  bpa_engie = bpa_engie[bpa_engie['CD_CONTA'].str.len() <= 7]

  # Reiniciar os indices
  bpa_engie.reset_index(inplace=True, drop=True)

  # Agrupamento em linhas
  bpa_engie_varios_anos = pd.concat([bpa_engie_varios_anos, bpa_engie])

bpa_engie_varios_anos = bpa_engie_varios_anos.pivot(index=['CD_CONTA','DS_CONTA'], columns='ANO', values='VL_CONTA')
bpa_engie_varios_anos
```

```python
bpa_engie_varios_anos['2019-2020'] = (bpa_engie_varios_anos[2020] / bpa_engie_varios_anos[2019] - 1) * 100
bpa_engie_varios_anos['2020-2021'] = (bpa_engie_varios_anos[2021] / bpa_engie_varios_anos[2020] - 1) * 100
```
